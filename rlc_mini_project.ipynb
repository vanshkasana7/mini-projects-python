PROJECT: RLC impedance vs frequency

This project simulates and visualizes the impedance of a Series RLC Circuit 
as a function of frequency using NumPy and Matplotlib.

Objective: Understand resonance and frequency response of RLC circuits.


```python
import numpy as np
import matplotlib.pyplot as plt

# Circuit parameters
R = 100
L = 200
C = 1e-60

# Frequency range
f = np.linspace(10, 10000, 1000)
omega = 2 * np.pi * f

# Impedance
Z = np.sqrt(R**2 + (omega * L - 1 / (omega * C))**2)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(f, Z, color='blue')
plt.title("Series RLC Circuit - Impedance vs Frequency")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Impedance (Ohms)")
plt.grid(True)
plt.savefig('output_RLC_impedance.png')
plt.show()
```

# Project: Underdamped Transient Response of a Series RLC Circuit

This project simulates the transient current response of a Series RLC Circuit 
in the underdamped condition using Python (NumPy & Matplotlib).

Objective: 
To visualize the oscillatory behavior of current when the damping ratio (Î¶) is less than 1, 
and understand how resistance, inductance, and capacitance values affect the system's response.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- PARAMETERS ---
R = 2          # Resistance in ohms
L = 1e-3      # Inductance in henrys
C = 1e-6      # Capacitance in farads

# --- DERIVED PARAMETERS ---
alpha = R / (2 * L)
omega_0 = 1 / np.sqrt(L * C)
zeta = alpha / omega_0

# Time array
t = np.linspace(0, 0.01, 1000)

# --- COMPUTE RESPONSE BASED ON DAMPING TYPE ---
if zeta < 1:  # Underdamped
    omega_d = omega_0 * np.sqrt(1 - zeta**2)
    i_t = np.exp(-alpha * t) * np.sin(omega_d * t)
    damping_type = "Underdamped"
elif zeta == 1:  # Critically damped
    i_t = t * np.exp(-alpha * t)
    damping_type = "Critically Damped"
else:  # Overdamped
    s1 = -alpha + np.sqrt(alpha**2 - omega_0**2)
    s2 = -alpha - np.sqrt(alpha**2 - omega_0**2)
    i_t = (np.exp(s1 * t) - np.exp(s2 * t))
    damping_type = "Overdamped"

# --- PLOT ---
plt.figure(figsize=(10, 5))
plt.plot(t, i_t, label=f'{damping_type} Response', color='blue')
plt.title(f'Series RLC Circuit Response ({damping_type})')
plt.xlabel('Time (s)')
plt.ylabel('Current (arbitrary units)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('output_RLC_damping.png')
plt.show() 

```
